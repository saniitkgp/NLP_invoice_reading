# Feature Engineering part

## Funtion to perform multicollinearity testing of features

```python
# multicollinearity testing
def perform_vif_testing(df):

    # Calculating VIF
    vif = pd.DataFrame()
    vif["variables"] = df.columns
    vif["VIF"] = [variance_inflation_factor(df.values, i) for i in range(df.shape[1])]

    return(vif)
```

## Convert string categorical data to numeric on dataframe using Ordinal Encoding

```python
def create_ordinal_data(data_df,encoder=None,is_training=True,cat_col_list=None):
#     import category_encoders as ce
    if is_training:
        if cat_col_list is None:
            cat_col_list =['veh_body','gender','area','agecat']
        encoder=ce.OrdinalEncoder(verbose=1,cols=cat_col_list,drop_invariant=True,return_df=True)
        data_df=encoder.fit_transform(data_df)
    else:
        data_df=encoder.transform(data_df)

    if is_training:
#         return encoder, data_df.to_numpy()
        return encoder, data_df
    else:
#         return data_df.to_numpy()
        return data_df
    
    
```

# Performing One Hot Encoding on dataframe

```python
def create_oneHot_data(data_df,encoder=None,is_training=True):
#     import category_encoders as ce
    if is_training:
        cat_col_list =['veh_body','gender','area','agecat']
        encoder=ce.OneHotEncoder(verbose=1,cols=cat_col_list,drop_invariant=True,return_df=True,use_cat_names=True)
        data_df=encoder.fit_transform(data_df)
    else:
        data_df=encoder.transform(data_df)
        
    drop_list =['veh_body_HBACK','gender_F','area_A','agecat_1.0']
    data_df = data_df.drop(columns=drop_list)  
    
    if is_training:
#         return encoder, data_df.to_numpy()
        return encoder, data_df
    else:
#         return data_df.to_numpy()
        return data_df
    
    
```

## Function to create R style formula

```python
def create_formula(df, target='numclaims',features=None):
    if features is None:
        features =[col for col in df.columns.tolist() if col != target]
    formula =target+' ~ '+features[0]
    for idx in range(1,len(features)):
        formula += ' + '+ features[idx]
    return formula
```
